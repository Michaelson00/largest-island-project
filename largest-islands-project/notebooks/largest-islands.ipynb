## Finding Islands with tropic climate

```python
tropical_islands=df.query("climate == 'tropics'").reset_index(drop=True)
```

```python
tropical_islands
```

## Sorting Island by area

```python
tropical_islands_sorted = tropical_islands.sort_values('area', ascending=False).reset_index(drop=True)
```

```python
tropical_islands_sorted
```

```python
top10_tropics = tropical_islands_sorted.head(10)
```

# Top 10 Tropical Islands


```python
top10_tropics
```

```python

x = top10_tropics['island']
y = top10_tropics['area']

plt.bar(x, y)
plt.xticks(rotation=90)
plt.xlabel('Island')
plt.ylabel('Area')
plt.title('Top 10 Largest Tropical Islands')
plt.show()
```

## What are largest islands in each region ?

## Finding largest island in each region

```python
region_largest_island = df.loc[df.groupby('region')['area'].idxmax()]
```

```python
region_largest_island
```

## Sorting largest Islands by decreasing area

```python
sorted_region_largest_island = region_largest_island.sort_values('area', ascending=False).reset_index(drop=True)
```

# Largest Island in Each Region

```python
sorted_region_largest_island
```

## Graph of Area by Region

```python

x = sorted_region_largest_island['region']
y = sorted_region_largest_island['area']

plt.bar(x, y)
plt.xticks(rotation=75)
plt.xlabel('Region')
plt.ylabel('Area')
plt.title('Largest Island in Each Region')
plt.show()
```

## Graph of Distribution of Island Areas

```python
x = df['area']

plt.hist(x, bins=20)
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.title('Distribution of Island Areas')
plt.show()
```

# Graph of Area vs Rank

```python
df_sorted = df.sort_values('rank', ascending=True)
x = df_sorted['rank']
y = df_sorted['area']
plt.plot(x, y)
plt.xlabel('Rank')
plt.ylabel('Area')
plt.yscale('log')
plt.title('Island Area vs Rank')
```

## A small number of islands dominate in size. After the first few ranks, the area drops sharply, and the majority of islands are significantly smaller.
